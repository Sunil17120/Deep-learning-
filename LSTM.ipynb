```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense,Dropout,LSTM
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
path='/content/IMDB Dataset.csv'
df=pd.read_csv(path)
```

```python
sentences=df['review'].values
labels=df['sentiment'].apply(lambda x: 1 if x=='positive' else 0).values
```

```python
tokenizer=Tokenizer(num_words=5000)
tokenizer.fit_on_texts(sentences)
sequences=tokenizer.texts_to_sequences(sentences)
```

```python
maxlen=200
x=pad_sequences(sequences,maxlen=maxlen)
y=np.array(labels)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=Sequential()
model.add(Embedding(input_dim=5000,output_dim=128))

model.add(LSTM(64,return_sequences=False
                    ,kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.7))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001),loss='binary_crossentropy',metrics=['accuracy'])
```

```python
early_stopping=EarlyStopping(monitor='val_loss',patience=3,restore_best_weights=True)
```

```python
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=128,callbacks=[early_stopping])
```

```python


_,test_accuracy=model.evaluate(x_test,y_test)
print(f'Test Accuracy: {test_accuracy*100:.2f}%')
```

```python
y_pred=(model.predict(x_test)>0.5).astype("int32")
print(f'accuracy score:{accuracy_score(y_test,y_pred):.4f}')
print(y_pred)
```
